```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(data.head())
```

```python
data.columns=["States","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Regions","longitude","latitude"]
```

```python
print(data)
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(8,7))
sns.heatmap(data.corr())
plt.show()
```

```python
data.columns=["States","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Regions","longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed",hue="Regions",data=data)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue="Regions",data=data)
plt.show()
```

```python
unemploment=data[["States","Regions","Estimated Unemployment Rate"]]
figure=px.sunburst(unemploment,path=["Regions","States"], values="Estimated Unemployment Rate",width=700, height=700,color_continuous_scale="RdY1Gn",title="Unemployment Rate in India")
# figure.write_html('Unemployment_Rate.html', auto_open=True)
figure.show()
```
